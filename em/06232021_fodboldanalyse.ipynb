### importer moduler og dataset

I dette afsnit henter vi datasettet fra SQL-databasen, som er indsamlet på siden encrypted.dk
Datasettet er hentet fra DBU (https://www.dbu.dk/nyheder/2021/maj/26-spillere-klar-til-em-for-danmark/) og er ikke blevet opdateret under EM-slutrunden.

```python
import pandas as pd #<- Pandas Dataframe

spillerdata = pd.read_csv('fodboldspillere.csv', sep = ';') # vores dataset fra SQL-databasen

print(spillerdata.to_string()) # se spillerdata som strenge
```

```python
# Hent en værdi af spillere midt i truppen med data-slicing
spillerdata[3:8]
```

```python
# Find minimum-værdier af spillerdataen
# Vi kan se at en er født 0000-00-00. Den bedste praksis er at ændre denne data i CSV-filen og IKKE i Jupyter
spillerdata.min(axis = 0)
```

## Antal kampe
I dette afsnit ser vi på analyse af, hvem der har spillet flest kampe

```python
# Variabel som indeholder spillernes kampe
kampe = spillerdata["Kampe"]
```

```python
# minimum antal kampe spillet af en spiller
kampe.min()
```

```python
# maximum
kampe.max()
```

```python
# gennemsnit antal kampe spillet
round(kampe.mean()) # round runder tallet
```

```python
# Kampe i hele truppen
spillerdata['Kampe']
```

```python
# her ser udelukker vi alle kolloner, bortset fra efternavn og kampe
spillerdata[['Efternavn', 'Kampe']]
```

```python
# Spillere med over 100 kampe
spillerdata[spillerdata['Kampe'] > 100]
```

## Opsummering af spillere med x antal kamp

```python
spillerdata.sort_values(by = "Kampe", ascending = False)
# Lidt forklaring på inplace: 
# https://stackoverflow.com/questions/45570984/in-pandas-is-inplace-true-considered-harmful-or-not
spillerdata
```

## Klubber med flest kampe

```python
gruppe_max = spillerdata[["Klub", "Kampe"]].groupby("Klub").sum()
gruppe_max
```

```python
# Gennemsnit af kampe
gruppe_gnms = spillerdata[["Klub", "Kampe"]].groupby("Klub").mean()
gruppe_gnms
```
